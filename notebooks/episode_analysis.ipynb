# Episode 数据分析

这个 notebook 演示如何读取 Guinsoo MuJoCo Platform 采集的 HDF5 + JSON 数据包。

```python
from pathlib import Path

import matplotlib.pyplot as plt

from guinsoo_mujoco.data import RunBrowser, default_runs_dir

runs_dir = Path(default_runs_dir())
artifacts = sorted(runs_dir.glob("*.h5"))

if artifacts:
    h5_path = artifacts[-1]
    json_path = h5_path.with_suffix(".json")
    episode = RunBrowser.open((h5_path, json_path))
    print(episode.metadata)
    print("qpos shape:", episode.qpos.shape)
    print("actuator_force shape:", episode.actuator_force.shape)

    joint_index = 0
    plt.figure(figsize=(10, 4))
    plt.plot(episode.time, episode.qpos[:, joint_index], label="qpos")
    plt.plot(episode.time, episode.target[:, joint_index], label="target")
    plt.plot(episode.time, episode.actuator_force[:, joint_index], label="actuator_force")
    plt.xlabel("time (s)")
    plt.ylabel("value")
    plt.legend()
    plt.title(episode.metadata.get("actuator_names", ["joint"])[joint_index])
    plt.show()
else:
    print(f"请先在 App 中记录 episode。默认目录：{runs_dir}")
```
