# Episode 数据分析

这个 notebook 演示如何读取 Guinsoo MuJoCo Platform 采集的 HDF5 + JSON 数据包。

```python
from pathlib import Path
from guinsoo_mujoco.data import RunBrowser

h5_path = Path("../data/runs/example.h5")
json_path = Path("../data/runs/example.json")

if h5_path.exists() and json_path.exists():
    episode = RunBrowser.open((h5_path, json_path))
    print(episode.metadata)
    print("qpos shape:", episode.qpos.shape)
else:
    print("请先在 App 中记录 episode，或修改 h5_path/json_path 指向已有数据。")
```
